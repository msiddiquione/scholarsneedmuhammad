## 1. Data Loading and Cleaning

We'll start by loading the dataset and cleaning messy data. This includes:
1.  Standardizing the year column
2.  Handling percentage columns (e.g., % Poverty, Economic Need Index) by converting strings like "Above 95%" to numeric values (0.95)
3.  Ensuring all data types are correct for analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Setting visualization style
sns.set_theme(style="whitegrid")

# ---------------------------------------------------------
# SIMULATED DATA LOADING (Replace with pd.read_csv('file.csv') in prod)
# ---------------------------------------------------------
from io import StringIO

csv_data = """
School Name,Year,Total Enrollment,Borough,Charter,Economic Need Index,% Poverty
PS 001,2018,500,Manhattan,N,75.0%,70.5%
PS 001,2019,510,Manhattan,N,76.2%,71.0%
PS 001,2020,490,Manhattan,N,78.5%,73.0%
PS 001,2021,480,Manhattan,N,80.0%,75.0%
PS 001,2022,475,Manhattan,N,82.1%,77.0%
Success Academy 1,2018,450,Bronx,Y,85.0%,82.0%
Success Academy 1,2019,520,Bronx,Y,84.0%,81.0%
Success Academy 1,2020,600,Bronx,Y,83.0%,80.0%
Success Academy 1,2021,680,Bronx,Y,82.0%,79.0%
Success Academy 1,2022,750,Bronx,Y,81.0%,Above 95%
Success Academy 2,2018,300,Brooklyn,Y,70.0%,65.0%
Success Academy 2,2019,350,Brooklyn,Y,68.0%,63.0%
Success Academy 2,2020,410,Brooklyn,Y,66.0%,60.0%
Success Academy 2,2021,460,Brooklyn,Y,65.0%,59.0%
Success Academy 2,2022,500,Brooklyn,Y,64.0%,58.0%
"""
df = pd.read_csv(StringIO(csv_data))

# ---------------------------------------------------------
# DATA CLEANING LOGIC
# ---------------------------------------------------------

# Function to handle specific percentage formatting issues
def clean_percentage_column(val):
    val_str = str(val).strip()
    if 'Above 95' in val_str:
        return 0.95
    if 'Below 5' in val_str:
        return 0.05

    # Remove symbols and convert
    clean_str = val_str.replace('%', '').replace(',', '')
    try:
        # Convert whole number percentage to decimal (e.g., 75.0 -> 0.75)
        return float(clean_str) / 100.0
    except ValueError:
        return np.nan

# Apply cleaning
perc_cols = ['Economic Need Index', '% Poverty']
for col in perc_cols:
    if col in df.columns:
        df[col] = df[col].apply(clean_percentage_column)

# Ensure Year is integer
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')

# Verify types
print("Data Cleaning Complete. Current Data Types:")
print(df.dtypes)
df.head()
```

## 2. Overall Market Analysis

**Prompt:** *What do you notice? What additional attributes would be beneficial?*

We analyze the data across three dimensions:
1.  Total Enrollment over time
2.  Enrollment breakdown by Borough
3.  Comparison between Charter vs. Non-Charter schools

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Total Enrollment Trend
annual_enrollment = df.groupby('Year')['Total Enrollment'].sum().reset_index()
sns.lineplot(data=annual_enrollment, x='Year', y='Total Enrollment', marker='o', ax=axes[0])
axes[0].set_title('Overall Market Enrollment Trend')

# 2. By Borough
borough_enrollment = df.groupby(['Year', 'Borough'])['Total Enrollment'].mean().reset_index()
sns.lineplot(data=borough_enrollment, x='Year', y='Total Enrollment', hue='Borough', ax=axes[1])
axes[1].set_title('Avg Enrollment by Borough')

# 3. Charter vs Non-Charter
charter_enrollment = df.groupby(['Year', 'Charter'])['Total Enrollment'].mean().reset_index()
sns.lineplot(data=charter_enrollment, x='Year', y='Total Enrollment', hue='Charter', ax=axes[2])
axes[2].set_title('Avg Enrollment: Charter (Y) vs Non-Charter (N)')

plt.tight_layout()
plt.show()
```

## 3. Projections: Success Academy (Linear vs. Polynomial)

**Prompt:** *Make a rough projection of enrollment at Success Academy Charter schools over the next 5 years.*

We will test two models:
1.  **Linear Regression (Baseline):** Assumes constant growth.
2.  **Polynomial Regression (Degree 2):** Assumes acceleration or deceleration.

We will compare them to make a strategic forecasting decision.

```python
# Filter for Success Academy only
sa_df = df[df['School Name'].str.contains('Success Academy')].copy()
sa_agg = sa_df.groupby('Year')['Total Enrollment'].sum().reset_index()

# Prepare Data
X = sa_agg[['Year']]
y = sa_agg['Total Enrollment']

# Future Years
future_years = np.array([[y] for y in range(2023, 2028)])

# --- Model 1: Linear ---
linear_model = LinearRegression()
linear_model.fit(X, y)
linear_pred = linear_model.predict(future_years)

# --- Model 2: Polynomial (Degree 2) ---
poly_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_model.fit(X, y)
poly_pred = poly_model.predict(future_years)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='black', label='Historical Data', s=100)
plt.plot(future_years, linear_pred, color='blue', linestyle='--', label='Linear Projection (Chosen)')
plt.plot(future_years, poly_pred, color='red', linestyle=':', label='Polynomial Projection (Rejected)')

plt.xlabel('Year')
plt.ylabel('Total Enrollment')
plt.title('Success Academy Enrollment Projection: Linear vs. Polynomial')
plt.legend()
plt.grid(True)
plt.show()
```
