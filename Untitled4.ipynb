```python
# --- 2. ANALYSIS FUNCTIONS ---
# Compute two growth rates (2021 and 2122), average them, and grow the 2023 value forward 5 years.
# If any of the recent years are missing or invalid, hold the projection flat at the 2023 level.

def calculate_projections(df):
    proj_years = [2024, 2025, 2026, 2027, 2028]

    for idx, row in df.iterrows():
        p21 = row['School_Age_Pop_2021']
        p22 = row['School_Age_Pop_2022']
        p23 = row['School_Age_Pop_2023']

        if p21 <= 0 or p22 <= 0 or p23 <= 0:
            for y in proj_years:
                df.at[idx, f'Proj_Pop_{y}'] = p23
            continue

        g1 = (p22 - p21) / p21
        g2 = (p23 - p22) / p22
        avg = (g1 + g2) / 2

        val = p23
        for y in proj_years:
            val = val * (1 + avg)
            df.at[idx, f'Proj_Pop_{y}'] = int(val)

    return df



def get_dashboard_requirements():
    return pd.DataFrame([
        ["Category", "Requirement"],
        ["User Types", "Executive Leadership: High-level aggregate views, YoY growth"],
        ["Key Metrics", "Enrollment Conversion: Applicants / School Age Pop"],
        ["Context", "Competitor Density & Housing Pipeline"],
        ["UI/UX", "Geospatial Filter (Borough -> Zip -> School)"]
    ])

def run_ai_analysis(data_df, reqs_df):
    if not model: return

    print("\n--- Generating AI Strategic Analysis ---")
    data_str = data_df.head(15).to_string()
    reqs_str = reqs_df.to_string()

    prompt = f"""
    You are a Senior Strategic Analyst for Success Academies.
    Review this dataset (Census trends, Projections, Competitors, Real Estate Pipeline).

    DATASET (Sample):
    {data_str}

    DASHBOARD REQUIREMENTS:
    {reqs_str}

    Provide a brief Executive Summary:
    1. Identify zip codes with alarming decline vs. high growth.
    2. Analyze 'Supply vs Demand' (Competitors vs. Housing).
    3. Comment on projection validity given Real Estate data.
    4. Recommend one "Killer Feature" for the dashboard.
    """

    try:
        response = model.generate_content(prompt)
        print("\n" + "="*40 + "\nGEMINI AI EXECUTIVE SUMMARY\n" + "="*40)
        print(response.text)
    except Exception as e:
        print(f"Error: {e}")

# --- 3. EXECUTION ---
if __name__ == "__main__":
    if os.path.exists(INPUT_FILE):
        print("--- Loading Data ---")
        df = pd.read_csv(INPUT_FILE)

        # Run Analysis
        df = calculate_projections(df)

        # Calculate Momentum
        df['Market_Momentum'] = (df['School_Age_Pop_2023'] - df['School_Age_Pop_2019']) / df['School_Age_Pop_2019']

        # Print Results
        print("\n--- FINAL RESULTS (Top 5) ---")
        pd.set_option('display.max_columns', None)
        pd.set_option('display.width', 1000)
        print(df.head().to_string())

        # Run AI
        run_ai_analysis(df, get_dashboard_requirements())
    else:
        print(f" File '{INPUT_FILE}' not found. Please run the data fetching script first.")
```

```python
import pandas as pd
import numpy as np
import os
import google.generativeai as genai
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. CONFIGURATION ---
INPUT_FILE = "nyc_school_market_data.csv"

# --- GEMINI SETUP ---
GOOGLE_API_KEY = None
model = None
try:
    from google.colab import userdata
    GOOGLE_API_KEY = userdata.get('collabda')
    if not GOOGLE_API_KEY: GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')

    if GOOGLE_API_KEY:
        genai.configure(api_key=GOOGLE_API_KEY)
        model = genai.GenerativeModel(model_name='models/gemini-3-pro-preview')
        print(" Gemini AI Configured.")
    else:
        print(" No API Key found. AI features disabled.")
except Exception as e:
    print(f"AI Setup Error: {e}")

# --- 2. ANALYSIS FUNCTIONS ---
def calculate_projections(df):
    """Generates 2024-2028 projections based on linear regression."""
    historical_years = np.array([2019, 2020, 2021, 2022, 2023])
    projection_years = np.array([2024, 2025, 2026, 2027, 2028])

    for idx, row in df.iterrows():
        y_values = [row.get(f'School_Age_Pop_{y}', 0) for y in historical_years]
        if sum(y_values) == 0: continue

        try:
            slope, intercept = np.polyfit(historical_years, y_values, 1)
            for p_year in projection_years:
                df.at[idx, f'Proj_Pop_{p_year}'] = max(0, round((slope * p_year) + intercept))
        except: pass
    return df

def visualize_results(df):
    """Creates key visualizations for the dashboard."""
    print("\n--- Generating Visualizations ---")

    # Ensure style is set without crashing if seaborn styles aren't available
    try:
        sns.set_theme(style="whitegrid")
    except:
        pass

    # 1. Top 5 Growth Trajectories
    # Sort by momentum to find top growers
    top_growth = df.sort_values('Market_Momentum', ascending=False).head(5)
    years = [2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028]

    plt.figure(figsize=(10, 6))
    for _, row in top_growth.iterrows():
        # Combine Hist + Proj
        hist = [row.get(f'School_Age_Pop_{y}', 0) for y in range(2019, 2024)]
        proj = [row.get(f'Proj_Pop_{y}', 0) for y in range(2024, 2029)]
        full_data = hist + proj
        # Handle cases where projection might have failed/be NaN
        full_data = [x if pd.notna(x) else 0 for x in full_data]

        plt.plot(years, full_data, marker='o', label=f"Zip {row['Zip_Code']}")

    plt.title('Top 5 Growth Zip Codes: Historical & Projected')
    plt.xlabel('Year')
    plt.ylabel('School Age Population')
    plt.axvline(x=2023, color='red', linestyle='--', label='Projection Start')
    plt.legend()
    plt.tight_layout()
    plt.savefig('growth_trends.png')
    print("  - Saved 'growth_trends.png'")
    plt.show() # Show in notebook

    # 2. Supply vs Demand Scatter
    if 'Private_School_Count' in df.columns and 'Planned_Housing_Units' in df.columns:
        plt.figure(figsize=(10, 6))
        # Filter out 0s for log scale or just better viz if needed, keeping simple for now
        sns.scatterplot(data=df, x='Private_School_Count', y='Planned_Housing_Units',
                        size='School_Age_Pop_2023', hue='Market_Momentum', sizes=(50, 500), alpha=0.7, palette='viridis')
        plt.title('Supply vs. Demand: Competitors vs. Housing Pipeline')
        plt.xlabel('Competitor Schools (Private)')
        plt.ylabel('Planned Housing Units')
        plt.axhline(0, color='grey', lw=1)
        plt.axvline(0, color='grey', lw=1)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.savefig('supply_demand_scatter.png')
        print("  - Saved 'supply_demand_scatter.png'")
        plt.show()

def get_dashboard_requirements():
    return pd.DataFrame([
        ["Category", "Requirement"],
        ["User Types", "Executive Leadership: High-level aggregate views, YoY growth"],
        ["Key Metrics", "Enrollment Conversion: Applicants / School Age Pop"],
        ["Context", "Competitor Density & Housing Pipeline"],
        ["UI/UX", "Geospatial Filter (Borough -> Zip -> School)"]
    ])

def run_ai_analysis(data_df, reqs_df):
    if not model: return

    print("\n--- Generating AI Strategic Analysis ---")
    data_str = data_df.head(15).to_string()
    reqs_str = reqs_df.to_string()

    prompt = f"""
    You are a Senior Strategic Analyst for Success Academies.
    Review this dataset (Census trends, Projections, Competitors, Real Estate Pipeline).

    DATASET (Sample):
    {data_str}

    DASHBOARD REQUIREMENTS:
    {reqs_str}

    Provide a brief Executive Summary:
    1. Identify zip codes with alarming decline vs. high growth.
    2. Analyze 'Supply vs Demand' (Competitors vs. Housing).
    3. Comment on projection validity given Real Estate data.
    4. Recommend one "Killer Feature" for the dashboard.
    """

    try:
        response = model.generate_content(prompt)
        print("\n" + "="*40 + "\nGEMINI AI EXECUTIVE SUMMARY\n" + "="*40)
        print(response.text)
    except Exception as e:
        print(f"Error: {e}")

# --- 3. EXECUTION ---
if __name__ == "__main__":
    if os.path.exists(INPUT_FILE):
        print("--- Loading Data ---")
        df = pd.read_csv(INPUT_FILE)

        # Run Analysis
        df = calculate_projections(df)

        # Calculate Momentum
        df['Market_Momentum'] = (df['School_Age_Pop_2023'] - df['School_Age_Pop_2019']) / df['School_Age_Pop_2019']

        # Print Results
        print("\n--- FINAL RESULTS (Top 5) ---")
        pd.set_option('display.max_columns', None)
        pd.set_option('display.width', 1000)
        print(df.head().to_string())

        # Run Visualization
        visualize_results(df)

        # Run AI
        run_ai_analysis(df, get_dashboard_requirements())
    else:
        print(f" File '{INPUT_FILE}' not found. Please run the data fetching script first.")
```

```python
import pandas as pd
import folium
import numpy as np

# --- NOTE: This is a standalone script for demonstration. ---
# --- When you run this, assume 'merged_df' is loaded from your analysis. ---

def create_market_dynamics_map(df):
    """
    Creates an interactive Folium map visualizing future market potential
    (Housing Pipeline size) and immediate competitive density (Private Schools color).

    Requires 'Lat', 'Lon', 'Planned_Housing_Units', and 'Private_School_Count' columns.
    """
    print("--- Generating Enhanced Market Dynamics Map ---")

    # Filter out rows with missing coordinates or zero units to prevent errors
    df_valid = df.dropna(subset=['Lat', 'Lon'])
    df_valid['Lat'] = pd.to_numeric(df_valid['Lat'], errors='coerce')
    df_valid['Lon'] = pd.to_numeric(df_valid['Lon'], errors='coerce')
    df_valid = df_valid.dropna(subset=['Lat', 'Lon'])

    if df_valid.empty:
        print("Error: No valid coordinates found to create map.")
        return None

    # Center map on the average location of the schools
    start_lat = df_valid['Lat'].mean()
    start_lon = df_valid['Lon'].mean()
    nyc_map = folium.Map(location=[start_lat, start_lon], zoom_start=11)

    # 1. Scaling for Marker Size (Housing Units)
    # Use log scaling to prevent very large outliers from dominating the map
    max_units = df_valid['Planned_Housing_Units'].max()
    scale_factor = 10 / np.log1p(max_units) if max_units > 0 else 5

    def get_radius(units):
        # Radius is proportional to the log of the number of planned units
        return np.log1p(units) * scale_factor

    # 2. Iterating and Adding Markers
    for _, row in df_valid.iterrows():
        # Metric 1: Private School Count determines color (Competitor Density)
        comp_count = row.get('Private_School_Count', 0)
        if comp_count >= 5:
            color = 'darkred'    # High competition
        elif comp_count >= 1:
            color = 'orange'     # Moderate competition
        else:
            color = 'darkgreen'  # Low competition

        # Metric 2: Planned Housing Units determines size (Future Demand)
        units = row.get('Planned_Housing_Units', 0)

        # Metric 3: Historical Growth (for context in popup)
        growth = row.get('Market_Momentum', np.nan)
        growth_str = f"{growth * 100:.1f}%" if pd.notna(growth) else "N/A"

        popup_html = f"""
        <b>DBN:</b> {row['DBN']}<br>
        <b>Zip Code:</b> {row['Zip_Code']}<br>
        <hr>
        <b>FUTURE DEMAND (Size):</b> {int(units)} Planned Units<br>
        <b>COMPETITION (Color):</b> {int(comp_count)} Private Schools<br>
        <b>POP. MOMENTUM (19-23):</b> {growth_str}
        """

        folium.CircleMarker(
            location=[row['Lat'], row['Lon']],
            radius=get_radius(units) + 5, # Minimum radius of 5
            popup=folium.Popup(popup_html, max_width=300),
            color=color,
            fill=True,
            fill_opacity=0.6,
            fill_color=color
        ).add_to(nyc_map)

    print("Map generated successfully. Check 'market_dynamics_map.html'.")
    return nyc_map

# -----------------------------------------------
# DEMONSTRATION EXECUTION
# -----------------------------------------------
if __name__ == "__main__":
    # Simulate a merged_df that contains the results from your analysis
    # (These are dummy values for demonstration only)
    sample_data = {
        'DBN': ['02M123', '15K001', '30Q501', '10X440', '31R080'],
        'Zip_Code': ['10010', '11222', '10031', '10467', '10301'],
        'Lat': [40.73837, 40.72145, 40.8185, 40.875, 40.64],
        'Lon': [-73.9813, -73.9533, -73.9499, -73.88, -74.10],
        'Planned_Housing_Units': [500, 1500, 50, 0, 200], # Pipeline size
        'Private_School_Count': [10, 3, 1, 0, 6],      # Competitor density
        'Market_Momentum': [-0.05, 0.15, -0.01, 0.08, 0.00] # Historical pop growth
    }
    simulated_df = pd.DataFrame(sample_data)

    # Call the map function
    map_object = create_market_dynamics_map(simulated_df)

    if map_object:
        # Save the interactive map to an HTML file
        #map_object.save("market_dynamics_map.html")
        # Display the map directly in the notebook output
        display(map_object)
```
